11-10 Ominaisvastus $\rho$ on $R \frac{A}{l}$ eli R=$\rho l/A$ https://fi.wikipedia.org/wiki/Ominaisvastus

Kirjan vastaus: a) 23 V b) 7.7 mA

```julia
N=1
A=30e3^2
B=52e-6
dt=20
dPhi=A*B*0.01
e=N*dPhi/dt
@printf("Indusoituva jännite on %.1f V.\n", e)
r=0.003/2
l=4*30e3
Aj=pi*r^2
R=l/Aj*0.18e-6
I=e/R
@printf("Virta on %.2e A.\n", I)
```

Kirjan vastaus: a) 8.0 µV b) 5.3 mA c) 0.27 µN (oikealle)

```julia
N=1
A=0.08^2
dt=0.5
dPhi=4e-6
e=N*dPhi/dt
@printf("a) Indusoituva jännite on %.1e V.\n", e)
R=1.5e-3
I=e/R
@printf("b) Virta on %.2e A.\n", I)
l=0.08 # Vain yksi sivu on magneettikentässä ja liikkuu sen suhteen
B=dPhi/A
Fm=B*I*l
@printf("c) Tarvittava voima on %.1e N.\n", Fm)
```

11-12 Jännite ajanhetkillä 0.25, 0.5, .75, 1.4
$u=-N\frac{d\Phi}{dt}$

Kirjan vastaus: b)1) 0.11 V, 2) 0 V 3) -0.11 V 4) 0.066 V

```julia
using Plots
using LaTeXStrings
x = 0:0.125/4:1.4
fx(x)=5e-6*cos(2*pi*x)
u(x)=3600*5e-6*sin(2*pi*x)*2*pi
@printf("""u(0.25)=%.2f V, u(0.5)=%.2f V,
  u(0.75)=%.2f V, u(1.4)=%.3f V\n""",
  u(0.25),u(0.5),u(0.75),u(1.4));
plot(x,fx.(x))
xlabel!(L"t")
ylabel!(L"\Phi")
```
